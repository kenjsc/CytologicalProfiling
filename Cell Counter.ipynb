```python
def main(pMap, edu, cyto, newFile = ''):
    ###takes in a plate map, two files of raw cp data from the csc.pbsci.ucsc.edu server that have been identified
    ###as either the cyto or the edu stain set, and writes the combined data to a file that has the total
    ###cell countrs for both plates and does some statistics
    p = PlateMap(pMap)
    #New plateMap object
    
    e = MXFile(edu)
    #New mxFile object
    
    c = MXFile(cyto)
    #New mxFile object
    
    cellAverage(p,e)
    cellAverage(p,c)
    standDev(p,e)
    standDev(p,c)
    
    newFile = open(newFile, 'wr')
    newFile.write("EdU Average," + str(e.meanUntreatedCellCount) + ',' + "EdUStdev" + ','+ str(e.stdeviationCellCount) + '\n')
    newFile.write("Cyto Average," + str(c.meanUntreatedCellCount) + ',' "CytoStdev" + ',' + str(c.stdeviationCellCount) + '\n')
    
    newFile.write("384 well, Compound Name,EdUQ1,EdUQ2,EdUQ3,EdUQ4,CytoQ1,CytoQ2,CytoQ3,CytoQ4,Sum EdU,Sum Cyto" + "\n")
    for x in p.wellMap:
        newFile.write(x + "," + p.wellMap[x] + "," + e.data[x] + "," + c.data[x] + "," + computeSum(e,x) + "," + computeSum(c,x) + "\n")
    newFile.close()
```

```python
 ###platemap object (pm), and a mxFile object (mx),and computes the average number of cells/well
def cellAverage(pm, mx):
    cells = 0
    blanks = 0
    for x in pm.wellMap:
        if bool(pm.wellMap[x] == '') or bool(pm.wellMap[x] == 'blank') or bool("DMSO" in pm.wellMap[x]):
            line = mx.data[x].split(',')
            blanks += 1
            for i in line:
                cells += eval(i)
    mx.meanUntreatedCellCount = float(cells/blanks)
```

```python
###summer takes a platemap object and a mxfile object and computes and returns the sum of each 4 quadrant well
def computeSum(mx, x):
    #pm = plateMap, mx = MXFile, x = dictionary key 'compound name'
    cells = 0
    for i in mx.data[x].split(','):
        cells += eval(i)
    return str(cells)
```

```python
###Standard Deviation of control wells
def standDev(pm, mx):
    diffCells = 0
    blanks = 0
    for x in pm.wellMap:
        if bool(pm.wellMap[x] == '') or bool('blank' in pm.wellMap[x]) or bool("DMSO" in pm.wellMap[x]):
            line = mx.data[x].split(',')
            blanks += 1
            cells = 0
            for i in line:
                cells += eval(i)
            diffCells += (mx.meanUntreatedCellCount - cells)*(mx.meanUntreatedCellCount - cells)
    mx.stdeviationCellCount = float(math.sqrt(diffCells/blanks))
```

```python
###dict version
class MXFile:
    def __init__(self, f):
        self.f = open(f,'U')
        self.headers = self.f.readline().split('\t')
        self.counter = self.headers.index('"Total Cells (MultiWaveScoring)"')
        self.wellName = self.headers.index('"Well Name"')
        self.data = {}
        self.meanUntreatedCellCount = 0
        self.stdeviationCellCount = 1.0
        for line in self.f:
            l = line.split('\t')
            if bool(l[self.wellName].strip('"') != '') & bool(l[self.counter] != ''):
                if str(l[self.wellName].strip('"')) in self.data:
                    self.data[l[self.wellName].strip('"')] += ',' + l[self.counter]
                else:
                    self.data.setdefault(l[self.wellName].strip('"'),l[self.counter])
```

```python
class PlateMap:
    def __init__(self, pm):
        self.p = open(pm, 'U')
        self.wellMap = {}
        self.headers = self.p.readline().strip().split(',')
        self.wells = self.headers.index('384 well')
        self.compound = self.headers.index('Molecule Name')
        for line in self.p:
            line = line.strip().split(',')
            if len(line) >= self.compound + 1:
                self.wellMap.setdefault(str(line[self.wells]))
                self.wellMap[str(line[self.wells])] = line[self.compound]
```

```python
###Array Version
class MXFile:
    def __init__(self, f):
        self.f = open(f,'U')
        self.headers = self.f.readline().split('\t')
        self.counter = self.headers.index('"Total Cells (MultiWaveScoring)"')
        self.wellName = self.headers.index('"Well Name"')
        self.data = []
        currentWell = ''
        combQuads = []
        for line in self.f:
            l = line.split('\t')
            if l[self.counter] != '':
                if str(l[self.wellName]) == currentWell:
                    combQuads.append(l[self.counter])
                else:
                    currentWell = l[self.wellName]
                    self.data.append(combQuads)
                    combQuads = [l[self.wellName],l[self.counter]]
        self.data.append(combQuads)
                
```
